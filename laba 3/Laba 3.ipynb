### Задание 1. Подсчет выборочных числовых характеристик
а) математическое ожидание, b) медиана, c) усеченное среднее (доля усеченных данных – аргумент функции), d) дисперсия, e) квантиль заданного порядка (уровень квантиля – аргумент функции), f) центральный и начальный момент заданного порядка (порядок – аргумент функции).

```python
import scipy
from scipy import stats
```

Центральный момент k порядка

```python
def central_moment(X, k):
    sum = 0
    mean = sample_moment(X, 1)
    for elem in X:
        sum+= (elem - mean) ** k
        
    return sum / len(X)
```

Начальный момент k порядка

```python
def sample_moment(X, k):
    sum = 0
    for elem in X:
        sum+= elem ** k
        
    return sum / len(X)
```

Математическое ожидание

```python
def mean(X):
    return sample_moment(X, 1)
```

Дисперсия

```python
def variance(X):
    return central_moment(X, 2)
```

Усеченное среднее.

Среднее усечённое представляет собой статистическую меру центральной тенденции, рассчитанную как среднее значение для имеющегося набора данных, из которого исключены k% наибольших и k% наименьших значений.

```python
def trim_mean(X, k):
    X_sorted = sorted(X)
    cnt_of_excluded = int(k * len(X_sorted))
    left_idx = cnt_of_excluded
    right_idx = len(X_sorted) - cnt_of_excluded
    return mean(X_sorted[left_idx:right_idx])
```

Квантиль k порядка.

Значение, которое заданная случайная величина не превышает с вероятностью k

```python
def quantile(X, k = 0.5):
    X_sorted = sorted(X)
    mid_left = int(k * len(X_sorted) - 1)
    mid_right = int(k*(len(X_sorted)))
    return (X_sorted[mid_left] + X_sorted[mid_right]) / 2
```

Медиана.

Так же квантиль порядка 0.5

```python
def median(X):
    return quantile(X, 0.5)
```

Проверить правильность работы функций, сравнив их выходы с выходами функций библиотек statistics/ numpy/ scipy на примере выборочных данных, полученных по результатам наблюдений за величиной ξ – видимость дороги в момент совершения аварии (Visibility)

```python
def compare_funcs(true, my, *args, name, eps = 1e-2):
    res_true = true(*args)
    res_my = my(*args)
    print("Встроенная функция: " + str(res_true))
    print("Моя функция: " + str(res_my))
    assert(abs(res_true- res_my) <= eps)
    print(name + " работает верно")
```

```python
data = data.dropna()
```

### Задание 2. Построить график, отражающий зависимость среднего значения видимости дороги от степени серьезности аварии.

```python
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("../data/02_car_accidents.xlsx")
```

```python
stats = [[], [], []]
severitys = np.arange(1, 5)
for severity in severitys:
    stats[0].append(data[data['Severity'] == severity]['Visibility(mi)'].mean())
    stats[1].append(data[data['Severity'] == severity]['Visibility(mi)'].median())
    stats[2].append(scipy.stats.trim_mean(data[data['Severity'] == severity]['Visibility(mi)'], 0.1))
    
fig, axes = plt.subplots(ncols=3, nrows=1, figsize = (20, 5))
fig.suptitle('Динамика изменения числовых признаков по severity ', fontsize=16)
names = ["mean", "median", "trunclated mean"]
for i in range(3):
    axes[i].set_xlabel("Severity")
    axes[i].set_ylabel(str(names[i]) + ' of visibility')
    sns.lineplot(severitys, stats[i], ax=axes[i], color = 'red')
```

Интуитивно мне казалось, что чем сильнее авария, тем меньше была видимость в тот момент. Однако исходя из полученных графиков видно, что видимость по мере роста серьезности не только не уменьшается, а даже растет! Получается, что аварии со степенью серьезности 1,2,3 не были вызваны ухудшением видимости. Однако для самой опасной степени серьезности 4 видимость и в правду меньше.

Заметим, что график математического ожидания (mean) немного отличается от графика усеченного среднего (trunclated mean) с параметром 0.1. Это видно в точках severity == 2 и severity == 3. Это означает, что в выборке есть немного выбросов для severity со значениями 2 и 3.

### Построить оценки функций распределения скорости ветра...

```python
quantiles = np.array([0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
data_stats = np.empty((4, quantiles.shape[0]))
for severity in range(4):
    data_stats[severity] = data[data['Severity'] == severity + 1]['Wind_Speed(mph)'].quantile(quantiles)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (20, 10))
fig.suptitle('Оценка функции распределения скорости ветра по квантилям')
for i in range(4):
    axes[i//2][i%2].set_xlabel("Wind speed")
    axes[i//2][i%2].set_title("Wind speed distribution, severity = " + str(i + 1))
    sns.lineplot(data_stats[i], quantiles, ax=axes[i//2][i%2], color = 'green')
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize = (20, 10))
fig.suptitle('Качественная оценка функции распределения скорости ветра, построенная по набору квантилей')
for i in range(4):
    axes[i//2][i%2].set_xlabel("Wind speed")
    axes[i//2][i%2].set_title("Wind speed distribution, severity = " + str(i + 1))
    axes[i//2][i%2].hist(data_stats[i],  cumulative=True, fill = False,
                         density = True, histtype = 'step', label = 'Эмпирическая ф.р.')
    sns.lineplot(data_stats[i], quantiles, ax=axes[i//2][i%2], color = 'green', label = 'Оценка ф.р. по квантилям')
```

Коробчатая диаграма распределения подтверждает выдвинутый мной тезис выше о том, что для данных с severity == 2 и severity == 3 есть выбросы.

### 1. Загрузка данных из документов с помощью Pandas

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv(filepath_or_buffer = "../data/file1.csv", sep = ',' ,encoding="cp1251",
                  names = ['Субъект РФ', '2015', '2018'])
data1 = data1.dropna()
data1 = data1.reset_index()
data1 = data1.drop(axis = 1, columns = 'index')
data1
```

```python
data2 = pd.read_csv(filepath_or_buffer = "../data/file2.csv", sep = ',' ,encoding="cp1251",
                  names = ['Субъект РФ', '2016', '2017'])
data2 = data2.dropna()
data2 = data2.reset_index()
data2 = data2.drop(axis = 1, columns = 'index')
data2
```

Итоговый датасет:

```python
data = pd.merge(data1, data2, how="left", on=["Субъект РФ"])
cols = ['2015', '2018', '2016', '2017']

# Не понимаю как пофиксить
# data[cols] = data[cols].apply(lambda x: x.replace(',', '.'))

data['2015'] = data['2015'].apply(lambda x: x.replace(',', '.'))
data['2016'] = data['2016'].apply(lambda x: x.replace(',', '.'))
data['2017'] = data['2017'].apply(lambda x: x.replace(',', '.'))
data['2018'] = data['2018'].apply(lambda x: x.replace(',', '.'))

data[cols] = data[cols].astype(float)
print(data.dtypes)
data
```

### 2. Составить словарь (dictionary), ключом в котором является название субъекта РФ, а значением – список из четырех элементов: число образовательных организаций в 2015, 2016, 2017, 2018 годах. + примечания

```python
data_corrected = data.reindex(columns=['Субъект РФ', '2015', '2016', '2017', '2018'])
data_corrected = data_corrected.loc[(data_corrected['Субъект РФ'].str.find('федеральный округ') == -1)
                                   & (data_corrected['Субъект РФ'].str.find('Российская Федерация') == -1)]

data_corrected = data_corrected.reset_index()
data_corrected = data_corrected.drop(axis = 1, columns = 'index')
```

```python
dict_data = data_corrected.set_index('Субъект РФ').T.to_dict('list')
dict_data
```

### 3. В полученном словаре для каждого субъекта РФ добавить в список 2 дополнительных элемента: среднее количество образовательных организаций за 4 года и год, в котором было достигнуто максимальное количество образовательных организаций

```python
def mean(list):
    return (list[0] + list[1] + list[2] + list[3]) / 4

def top_index(list):
    list2 = []
    idx = 0
    for item in range(4):
        list2.append([list[item], idx])
        idx = idx + 1
        
    list2 = sorted(list2, reverse = True)
    return list2[0][1] + 2015    
```

```python
# Хотел сделать лямбду, но я слишком глупый для этого :(
for key, value in dict_data.items():
     value.append(mean(value))
        
dict_data
```

```python
for key, value in dict_data.items():
     value.append(top_index(value))
        
dict_data
```

### 4. С использованием библиотечного средства Counter составить словарь, в котором ключом является год из диапазона 2015-2018, а значением – количество субъектов РФ, в которых именно в этом году количество образовательных организаций было максимальным.

Нигде в интернете не нашел guide по использованию Counter по словарю так, как написано в постановке задачи. Поэтому сделаем Counter по pd.DataSeries

```python
from collections import Counter
```

```python
data_help = data_corrected.copy()
data_help['max_year'] = data_help[cols].idxmax(axis = 1)
data_help
```

```python
counts = Counter(data_help['max_year'])
print (counts)
```
